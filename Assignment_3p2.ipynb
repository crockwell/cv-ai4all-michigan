```python
from __future__ import print_function
from imutils.object_detection import non_max_suppression
from imutils import paths
import numpy as np
import os
import imutils
import cv2
```

```python
dirs = ['Images/CityScapes', 'Images/UT-Interaction']
paths = []
for dir in dirs:
    files = os.listdir(dir)
    for file in files:
        paths.append(os.path.join(dir,file))
```

Did you know that OpenCV has built-in methods to perform pedestrian detection?

OpenCV ships with a pre-trained HOG + Linear SVM model that can be used to perform pedestrian detection in both images and video streams.

Source: https://www.pyimagesearch.com/2015/11/09/pedestrian-detection-opencv/

```python
# initialize the HOG descriptor/person detector
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
```

```python
from skimage.feature import hog as sk_hog
from skimage import data, exposure
import matplotlib.pyplot as plt
   
def plot_hog(img):
    plt.clf()
    
    _, hog_image = sk_hog(img, orientations=8, pixels_per_cell=(16, 16),
                    cells_per_block=(1, 1), visualize=True, multichannel=True)
    # Rescale histogram for better display
    hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0, 10))
    
    plt.imshow(hog_image_rescaled, cmap=plt.cm.gray)
    plt.show()
```

```python
def detect_people(image):
    image = imutils.resize(image, width=min(600, image.shape[1]))
    orig = image.copy()
    orig2 = image.copy()
    
    # detect people in the image
    (rects, weights) = hog.detectMultiScale(image, winStride=(4, 4),
                                            padding=(8, 8), scale=1.05)
    
    rects = np.array([[x, y, x + w, y + h] for (x, y, w, h) in rects])
    pick = non_max_suppression(rects, probs=None, overlapThresh=0.65)
 
    # draw the final bounding boxes
    for (xA, yA, xB, yB) in pick:
        cv2.rectangle(image, (xA, yA), (xB, yB), (0, 255, 0), 2)
 

    # show the output images
    cv2.imshow('Output Image', image)
    plt.show()
    
    # draw the original bounding boxes
    for (x, y, w, h) in rects:
        plot_hog(orig2[y:h, x:w])
        plt.imshow(orig2[y:h, x:w])
        plt.show()
```

```python
for imagePath in paths:
    _, fname = os.path.split(imagePath)
    print(imagePath)
    if fname[0] == '.':
        continue
    img = cv2.imread(imagePath)
    detect_people(img)
```
