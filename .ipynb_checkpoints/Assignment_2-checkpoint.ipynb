# Assignment 2: Template Matching

```python
%matplotlib inline 
import matplotlib.pyplot as plt
import numpy as np
import cv2 
```

First, we're going to apply the moving average (box) filter.

Step 1: Read in the "flower.png" image and display it:

```python
# Load a color image
polka = cv2.imread('Images/polka.png')
polka = cv2.cvtColor(polka, cv2.COLOR_RGB2GRAY )
polka = cv2.GaussianBlur(polka, (5,5), 3)

plt.imshow(polka)
plt.show()

pattern = polka[130:280, 65:205]
plt.imshow(pattern)
plt.show()

res = cv2.matchTemplate(polka, pattern, cv2.TM_CCORR)
plt.imshow(res)
plt.show()
```

```python
# Load a color image
img = cv2.imread('Images/polka_puppy.png')
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY )

plt.imshow(img)
plt.show()


res = cv2.matchTemplate(img, pattern, cv2.TM_CCOEFF)
plt.imshow(res)
plt.show()
```

```python
# Load a color image
beatles = cv2.imread('Images/beatles.jpg')
beatles = cv2.cvtColor(beatles, cv2.COLOR_RGB2GRAY )
beatles = cv2.GaussianBlur(beatles, (5,5), 3)

plt.imshow(beatles)
plt.show()

face = cv2.imread('Images/one_beatle.png')
face = cv2.cvtColor(face, cv2.COLOR_RGB2GRAY )
plt.imshow(face)
plt.show()


res = cv2.matchTemplate(beatles, face, cv2.TM_CCOEFF)
plt.imshow(res)
plt.show()
```
